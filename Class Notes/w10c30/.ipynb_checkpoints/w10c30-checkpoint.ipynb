### usa.gov data from bit.ly

In 2011, URL shortening service bit.ly partnered with the United States government website usa.gov to provide a feed of anonymous data gathered from users who shorten links ending with .gov or .mil . 

This project was maintained at https://github.com/usagov/1.USA.gov-Data, but was shutdown on July 2016. 


```python
open("usagov_bitly.txt").readlines()[0:2]
```

From this dump of the first 2 lines, we can see that this looks like it's in json format. 

As we have already soon, Python offers a number of modules that allow us to easily manage json data. We will use the json library. 

```python
import json

records = [json.loads(line) for line in open("usagov_bitly.txt")]

records[1:3]
```

```python
records[0]['tz']
```

```python
time_zones = [rec['tz'] for rec in records]
```

```python
for rec in records:
    print(rec['tz'])
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

Now, let's say we want to analyze where the traffic to .gov sites is coming from.

Let's use timezone as a proxy indicator for location and count the occurances of a given timezone.

```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
tz_counts = get_counts(time_zones)
```

```python
tz_counts['America/New_York']
```

Let's try counting timezones

```python
def top_counts(count_dict, n=10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort(reverse=True)
    return value_key_pairs[:n]
```

```python
top_counts(tz_counts)
```

## Counting with Pandas

As we have already seen, the main data structure in Pandas is a DataFrame. Let's take our records data structure (a list of dictionaries) and load this into a Pandas DataFrame

```python
from pandas import DataFrame, Series
import pandas as pd

df = DataFrame(records)

df
```

NOTE: When we think of "missing data", there are two instances that we find. The first is that there is a field identified, but there is not value in this field. This is blank data. But, there are times when there is no variable as all, and in such instances, we interpret this as NA (not available).

In Python Data Frames, we see "blank" values as simply field with nothin in them. For missing values, we see these represented as NaN (Not a number).

So, for any of our records that have a tz field but have nothing in it, the field will simply display as blank -- nothing in it. On the other hand, any records that don't have a tz field will display a NaN. 


```python
tz_counts = df['tz'].value_counts()
tz_counts[:10]
```

```python
clean_tz = df['tz'].fillna('Missing')
clean_tz[clean_tz==''] = "Unkown"
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

```python
%matplotlib inline
tz_counts[:10].plot(kind='barh', rot=0)
```

```python
from bokeh.charts import Bar
from bokeh.io import output_notebook, show

output_notebook()
type(tz_counts)

p = Bar(tz_counts[:10],)
show(p)
```

```python
results = Series([x.split()[0] for x in df.a.dropna()])
results[:5]
```

```python
import numpy as np
osf = df[df.a.notnull()]
os = np.where(osf['a'].str.contains('Windows'), 'Windows', 'Not Windows')
print(os)
```

```python
by_tz_os = osf.groupby(['tz', os])
by_tz_os.describe()
```

```python
by_tz_os.head(5)
```

```python
agg_counts = by_tz_os.size()
agg_counts
```

```python
agg_counts.unstack(0)
```

```python
agg_counts.unstack(1)
```

```python
agg_counts.unstack(0).unstack(0)
```

```python
agg_counts.unstack().fillna(0)
```
